```python
import pandas as pd
import plotly.express as px
```

```python
## Read in ultra rankings table
ultra_rankings = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-10-26/ultra_rankings.csv")
```

```python
#Read in race table
race = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-10-26/race.csv")
```

```python
#Check out ultra rankings table
ultra_rankings.head(5)
```

```python
#Check out race table
race.head(5)
```

```python
#Inner join the two tables on race year id
results = pd.merge(ultra_rankings, race, how = 'inner', on = 'race_year_id')
```

```python
#Check out joined tables
results.head(5)
```

```python
#Breakdown finishers by gender
results['gender'].value_counts()
```

```python
#See which participants have a race distance of 0 miles
distance = results['distance'] == 0.0
results[distance]
```

```python
#Get a count for how many participants have a race distance of 0 miles
results[distance].shape 
```

```python
#See what races frequently have a race distance of 0 miles
results[distance].groupby('event').size()
```

```python
#I realized that some of the events might be categorized as two different events, despite being the same one, because of being entered as all caps or lower case, so I converted the events column to all upper case
results['event'] = results['event'].str.upper()
results[distance].groupby('event').size()
```

```python
#Update Leadville manually 
results.loc[results['city'] == 'Leadville', 'distance'] = 100.0
distance2 = results['distance'] == 0.0
results[distance2].shape
```
